## For Singapore DATA only

```python
# To filter only for Singapore Data
#Replace 'Singapore' for 'United States' for US Data, change graph accordingly

filtered_df = df[(df['country'] == 'Singapore') & (df['year'].between(2004, 2024))]
filtered_df
```

```python
# Replace NaN values with the mean of the column
filtered_df['Inflation, GDP deflator (annual %)'] = filtered_df['Inflation, GDP deflator (annual %)'].fillna(filtered_df['Inflation, GDP deflator (annual %)'].mean())
```

```python
# Plotting the time series for inflation, GDP deflator (annual %)
plt.figure(figsize=(10,6))
plt.plot(filtered_df['year'], filtered_df['Inflation, GDP deflator (annual %)'], marker='o', linestyle='-', color='b')
plt.title('Inflation (GDP deflator) in Singapore Over Time (Till 2021)')
plt.xlabel('Year')
plt.ylabel('Inflation, GDP deflator (annual %)')
plt.grid(True)
plt.show()
```

```python
# Plotting the time series for Inflation,
plt.figure(figsize=(10,6))
plt.plot(filtered_df['year'], filtered_df['Inflation, consumer prices (annual %)'], marker='o', linestyle='-', color='g')
plt.title('Inflation (Consumer Prices) in Singapore Over Time (Till 2021)')
plt.xlabel('Year')
plt.ylabel('Inflation, consumer prices (annual %)')
plt.grid(True)
plt.show()
```

```python
# For missing data in the unemployment column
filtered_df['Unemployment, total (% of total labor force) (national estimate)'] = filtered_df['Unemployment, total (% of total labor force) (national estimate)'].fillna(filtered_df['Unemployment, total (% of total labor force) (national estimate)'].mean())
```

```python
plt.figure(figsize=(10,6))
plt.plot(filtered_df['year'], filtered_df['Unemployment, total (% of total labor force) (national estimate)'], marker='o', linestyle='-', color='b')
plt.title('Unemployment Rate (National Estimate) Over Time in Singapore (Till 2021)')
plt.xlabel('Year')
plt.ylabel('Unemployment, total (% of total labor force)')
plt.grid(True)
plt.show()
```

## Combination of sentiments

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample code to read your data into a DataFrame
# Replace 'your_data.csv' with your actual file path
df = pd.read_csv('labelled_corpus.csv')

# Ensure the 'date' column is in datetime format, with UTC handling for mixed time zones
df['date'] = pd.to_datetime(df['date'], dayfirst=True, utc=True, errors='coerce')

# Extract the year after ensuring datetime conversion
df['year'] = df['date'].dt.year

# Filter data for 'Negative' sentiment
negative_df = df[df['sentiment'] == 'Negative']

# Count the number of negative sentiments per year
negative_count_by_year = negative_df.groupby('year').size()

# Plotting
plt.figure(figsize=(10, 6))
negative_count_by_year.plot(kind='bar', color='red')
plt.title('Number of Negative Sentiments by Year')
plt.xlabel('Year')
plt.ylabel('Count of Negative Sentiments')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample code to read your data into a DataFrame
# Replace 'your_data.csv' with the path to your actual file
df = pd.read_csv('labelled_corpus.csv')

# Convert the 'date' column to datetime format and extract the year
df['date'] = pd.to_datetime(df['date'], dayfirst=True, utc=True, errors='coerce')
df['year'] = df['date'].dt.year

# Group by year and sentiment, and count the entries
sentiment_counts = df.groupby(['year', 'sentiment']).size().unstack(fill_value=0)

# Plotting without stacking
plt.figure(figsize=(12, 8))
sentiment_counts.plot(kind='bar', stacked=False, color=['green', 'blue', 'red'], figsize=(12, 6))
plt.title('Sentiment Counts by Year')
plt.xlabel('Year')
plt.ylabel('Count of Sentiments')
plt.xticks(rotation=45)
plt.legend(title="Sentiment")
plt.show()
```

```python
# Assume the uploaded file has a name 'your_file.csv'
df_mental = pd.read_csv('mentalhealthtrend.csv')
```

```python
df_mental.head()
```

```python
print(df_mental.columns)
```

```python
# Convert 'Month' to datetime and extract the year
df_mental['Month'] = pd.to_datetime(df_mental['Month'], format='%Y-%m')
df_mental['Year'] = df_mental['Month'].dt.year

# Filter for data from 2010 onwards
df_mental = df_mental[df_mental['Year'] >= 2010]

# Group by year and calculate the average
yearly_data = df_mental.groupby('Year')['mental health: (Worldwide)'].mean()

# Output the result
print(yearly_data)
```

```python
# Plot the yearly data as a time series chart
plt.figure(figsize=(10, 6))
plt.plot(yearly_data.index, yearly_data.values, marker='o', linestyle='-', color='b')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Average Mental Health Score (Interest)')
plt.title('Average Mental Health Interest (Worldwide) by Year')
plt.grid(True)

# Show the chart
plt.show()
```

```python
df_mental_region = pd.read_csv('mentalhealthtrend_byregion.csv')
```

```python
df_mental_region.head()
```

```python
# Use Plotly's choropleth to visualize mental health scores on a world map
fig = px.choropleth(
    df_mental_region,
    locations='Country',
    locationmode='country names',  # Match country names directly
    color='mental health',
    hover_name='Country',
    color_continuous_scale='Viridis',
    title="Mental Health Interest Scores by Country (2004-2024)"
)

# Add annotation for the note
fig.add_annotation(
    text="Note: Values are calculated on a scale from 0 to 100, where 100 is the location with the most popularity as a fraction of total searches in that location. "
         "A value of 50 indicates a location which is half as popular. A value of 0 indicates a location where there was not enough data for this term.",
    xref="paper", yref="paper",
    x=0.5, y=-0.1,  # Position the text below the map
    showarrow=False,
    font=dict(size=10),
    align="center"
)

# Adjust layout to make room for the note
fig.update_layout(
    margin={"r":0, "t":40, "l":0, "b":100}  # Adds space for the note at the bottom
)

# Show the interactive map
fig.show()
```

```python
# Sort data by mental health score in descending order


# Select the top 5 countries
top_5 = df_mental_region[:5]

print(top_5)
# Extract countries and mental health scores

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_5['Country'], top_5['mental health'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Mental Health Index')
plt.title('Mental Health Interest Score by Country (Top 5)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```
